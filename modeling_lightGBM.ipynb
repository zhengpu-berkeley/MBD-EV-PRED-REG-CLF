```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set pandas to display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('new_222_data_train_decorr_imputed.csv')
df_test = pd.read_csv('new_222_data_test_decorr_imputed.csv')
df.sample(3)
```

```python
X = df.loc[:, df.columns != 'microbusiness_density']
y = df['microbusiness_density']
y = np.log(y)

X_test = df_test.loc[:, df_test.columns != 'microbusiness_density']
y_test = df_test['microbusiness_density']
y_test = np.log(y_test)
```

```python
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# get rid of special characters in column names, only keep alphanumeric and underscore
X.columns = X.columns.str.replace('[^A-Za-z0-9_]+', '')
X_test.columns = X_test.columns.str.replace('[^A-Za-z0-9_]+', '')
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor
```

```python
param_grid = {
    'num_leaves': [5, 10, 15], 
    'max_depth': [5, 10],
    "learning_rate": [0.1, 0.05, 0.01],
    "n_estimators": [100, 500, 1000],
    "random_state": [42]
}
lgbm = LGBMRegressor()
grid_search = GridSearchCV(lgbm, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_search.fit(X, y)

print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
param_grid = {
    'num_leaves': [3, 5, 7], 
    'max_depth': [3, 5, 7],
    "learning_rate": [0.07, 0.05, 0.03],
    "n_estimators": [300, 500, 700],
    "random_state": [42]
}
lgbm = LGBMRegressor()
grid_search = GridSearchCV(lgbm, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_search.fit(X, y)

print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
param_grid = {
    'num_leaves': [2, 3, 4], 
    'max_depth': [1, 2, 3],
    "learning_rate": [0.06, 0.065, 0.07],
    "n_estimators": [650, 700, 750],
    "random_state": [42]
}
lgbm = LGBMRegressor()
grid_search = GridSearchCV(lgbm, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_search.fit(X, y)

print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
best_param = {'learning_rate': 0.05, 'max_depth': 10, 'n_estimators': 500, 'num_leaves': 10, 'random_state': 42}
lgbm = LGBMRegressor(**best_param)

# get cross validation rmse
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lgbm, X, y, cv=5, scoring='neg_mean_squared_error')
print('Cross validation RMSE: ', np.sqrt(-scores).mean(), '±' ,np.sqrt(-scores).std())
```

```python
# test set rmse
lgbm.fit(X, y)
y_pred = lgbm.predict(X_test)
from sklearn.metrics import mean_squared_error
print('Test set RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

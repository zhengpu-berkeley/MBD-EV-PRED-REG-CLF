```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set pandas to display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('new_222_data_train_norm_decorr.csv')
test_df = pd.read_csv('new_222_data_test_norm_decorr.csv')

df.sample(3)
```

# Target : log(MBD)

```python
y = df['microbusiness_density']
y = np.log(y)
y_bar = y.mean()
```

```python
y_bar_rmse = np.sqrt(np.mean((y - y_bar)**2))
y_bar_rmse
```

```python
y_test = test_df['microbusiness_density']
y_test = np.log(y_test)

y_bar_rmse_test = np.sqrt(np.mean((y_test - y_bar)**2))
y_bar_rmse_test
```

```python
y_leq_10 = y[df['microbusiness_density'] < 10]
```

```python
y_leq_10_bar = y_leq_10.mean()
```

```python
y_leq_10_bar_rmse = np.sqrt(np.mean((y - y_leq_10_bar)**2))
y_leq_10_bar_rmse
```

```python
y_leq_10_bar_rmse_test = np.sqrt(np.mean((y_test - y_leq_10_bar)**2))
y_leq_10_bar_rmse_test
```

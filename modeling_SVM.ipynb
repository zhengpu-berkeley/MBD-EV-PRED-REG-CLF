```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set pandas to display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('new_222_data_train_norm_decorr.csv')
df_test = pd.read_csv('new_222_data_test_norm_decorr.csv')
df.sample(3)

df.sample(4)
```

```python
X = df.loc[:, df.columns != 'microbusiness_density']
y = df['microbusiness_density']
y = np.log(y)

X_test = df_test.loc[:, df_test.columns != 'microbusiness_density']
y_test = df_test['microbusiness_density']
y_test = np.log(y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
# define the parameter values that should be searched
C_range = [0.01, 0.1, 1, 10, 100]
gamma_range = [0.01, 0.1, 1, 10, 100]

# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(gamma=gamma_range, C=C_range, kernel=['rbf'])

# instantiate and fit the grid
grid = GridSearchCV(SVR(), param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid.fit(X, y)

# best parameters
print(grid.best_params_)
```

```python
# define the parameter values that should be searched
C_range = [0.5, 0.07, 1, 4, 7]
gamma_range = [0.0001, 0.01, 0.1, 0.3, 0.7]

# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(gamma=gamma_range, C=C_range, kernel=['rbf'])

# instantiate and fit the grid
grid = GridSearchCV(SVR(), param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid.fit(X, y)

# best parameters
print(grid.best_params_)
```

```python
# define the parameter values that should be searched
C_range = np.logspace(-1,1,20)
gamma_range = np.logspace(-3,-1,20)

# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(gamma=gamma_range, C=C_range, kernel=['rbf'])

# instantiate and fit the grid
grid = GridSearchCV(SVR(), param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid.fit(X, y)

# best parameters
print(grid.best_params_)
```

```python
best_params = {'C': 1.438449888287663, 'gamma': 0.003359818286283781, 'kernel': 'rbf'}

# instantiate model with best parameters
svr = SVR(**best_params)

# cross validation rmse
from sklearn.model_selection import cross_val_score
scores = cross_val_score(svr, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)
print(rmse_scores)
print(rmse_scores.mean())
```

```python
# test rmse
from sklearn.metrics import mean_squared_error
svr.fit(X, y)
y_pred = svr.predict(X_test)
svr_mse = mean_squared_error(y_test, y_pred)
svr_rmse = np.sqrt(svr_mse)
print(svr_rmse)
```
